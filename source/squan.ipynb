```python
from ipycanvas import Canvas
```

```python
colors = {
    "U": "#000000",
    "D": "#FFFFFF",
    "F": "#FF0000",
    "B": "#FFA500",
    "R": "#00FF00",
    "L": "#0049FF"
}
```

```python
canvas = Canvas()
canvas.scale(2)

def draw_square(x, y, size):
    canvas.begin_path()
    canvas.move_to(x, y)

    canvas.line_to(x + size, y)
    canvas.line_to(x + size, y + size)
    canvas.line_to(x, y + size)
    canvas.line_to(x, y)

    canvas.stroke()
    canvas.close_path()

draw_square(100, 100, 50)

canvas
```
